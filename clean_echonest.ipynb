```python
import pandas as pd
import numpy as np
```

```python
echonest = pd.read_csv('echonest.csv')
```

```python
echonest['Unnamed: 0'].iloc[0] = echonest['Unnamed: 0'].iloc[2]
echonest['Unnamed: 0'].iloc[1] = ''

echonest.columns = echonest.iloc[0]+ '_'+ echonest.iloc[1]

echonest.columns.values[0] = 'track_id'

echonest = echonest[3:]
echonest = echonest.reset_index(drop=True)
```
